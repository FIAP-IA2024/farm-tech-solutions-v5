5. Encontrando Agrupamentos nos dados (Clusters)

```python
from sklearn.cluster import KMeans

# Selecionando apenas as variáveis numéricas para a clusterização
X_cluster = df_numerico[['Precipitation (mm day-1)', 'Temperature at 2 Meters (C)', 'Yield']]

# Definindo o número de clusters (vamos testar 3 agrupamentos)
kmeans = KMeans(n_clusters=3, random_state=42)
df_numerico["Cluster"] = kmeans.fit_predict(X_cluster)

# Visualizando os clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df_numerico["Temperatura a 2 metros (ºC)"], 
                y=df_numerico["Yield"], 
                hue=df_numerico["Cluster"], 
                palette="viridis")
plt.title("Clusters de Produtividade das Safras")
plt.show()
```
